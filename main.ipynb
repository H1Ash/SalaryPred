```python
df = df[["Country", "EdLevel", "YearsCodePro",
         "Employment", "ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df = df.rename({"YearsCodePro": "CodingExperiance"}, axis=1)
```

```python
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment", axis=1)
```

```python
df.Country.value_counts()
```

```python
def shorten_categories(cat, cutoff):
    categorial_map = {}
    for i in range(len(cat)):
        if cat.values[i] >= cutoff:
            categorial_map[cat.index[i]] = cat.index[i]
        else:
            categorial_map[cat.index[i]] = "other"
    return categorial_map
```

```python
Country_map = shorten_categories(df.Country.value_counts(), 500)
df.Country = df.Country.map(Country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1, 1)
df.boxplot("Salary", "Country")
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df.Salary <= 175000]
df = df[df.Salary >= 10000]
df = df[df.Country != "other"]
```

```python
df.CodingExperiance.unique()
```

```python
def clean_exp(x):
    if x == "Less than 1 year":
        return 0.5
    if x == "More than 50 years":
        return 50
    return float(x)


df.CodingExperiance = df.CodingExperiance.apply(clean_exp)
```

```python
df
```

```python
df.EdLevel.unique()
```

```python
def clean_ed(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor'
    if "Master’s degree" in x:
        return 'Masters'
    if 'Professional degree' in x or 'Other doctoral degree' in x:
        return 'PostGrad'
    return "less than bachelor's"


df.EdLevel = df.EdLevel.apply(clean_ed)
df.EdLevel.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le_ed = LabelEncoder()
df.EdLevel = le_ed.fit_transform(df.EdLevel)
df.EdLevel.unique()
```

```python
le_cnt = LabelEncoder()
df.Country = le_cnt.fit_transform(df.Country)
df.Country.unique()
```
